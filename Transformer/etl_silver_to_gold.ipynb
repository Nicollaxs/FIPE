```python
import pandas as pd
```

```python
PATH_SILVER = '../Data Layer/silver/fipe_silver.parquet'
DIR_GOLD = '../Data Layer/gold/'

# 1. Carregar a Silver
df = pd.read_parquet(PATH_SILVER)

# --- FATO 1: EVOLUÇÃO HISTÓRICA DE PREÇOS ---
# Agregamos por Ano de Referência e Marca para ver a tendência
fato_evolucao_precos = df.groupby(['ano_referencia', 'nome_marca'], observed=True).agg(
    preco_medio=('valor_reais', 'mean'),
    preco_mediano=('valor_reais', 'median'),
    volatilidade_preco=('valor_reais', 'std'),
    quantidade_registros=('valor_reais', 'count')
).reset_index()

fato_evolucao_precos.to_parquet('../Data Layer/gold/fato_evolucao_precos.parquet')
```

```python
fato_evolucao_precos.head()
```

```python
fato_depreciacao = df.groupby(['nome_modelo', 'idade_veiculo'], observed=True).agg(
    valor_medio_por_idade=('valor_reais', 'mean')
).reset_index()

fato_depreciacao.to_parquet('../Data Layer/gold/fato_depreciacao.parquet')
```

```python
fato_depreciacao.head()
```

```python
dim_veiculos = df[['nome_marca','nome_modelo','tipo_veiculo']].drop_duplicates().reset_index(drop = True)
dim_veiculos.to_parquet('../Data Layer/gold/dim_veiculos.parquet')
```

```python
datas = pd.date_range(start='2010-01-01', end='2026-12-31', freq='D')

dim_calendario = pd.DataFrame(datas, columns=['data'])

dim_calendario['ano'] = dim_calendario['data'].dt.year
dim_calendario['mes_num'] = dim_calendario['data'].dt.month
dim_calendario['mes_nome'] = dim_calendario['data'].dt.strftime('%B')
dim_calendario['trimestre'] = dim_calendario['data'].dt.quarter
dim_calendario['ano_mes'] = dim_calendario['data'].dt.strftime('%Y-%m')

dim_calendario.to_parquet('../Data Layer/gold/dim_calendario.parquet', index=False)

print("Dimensão Calendário criada com sucesso!")
```
